```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1, 2, 3, 4])  # x = [0, 1, 2, 3], y = [1, 2, 3, 4]
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'o')  # x = [1, 2, 3, 4], y = [1, 4, 9, 16]
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'o') 
plt.axis([0, 6, 0, 10])  # x = [0, 6], y = [0, 10] に軸の範囲を設定
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', label='y=x')
plt.plot(t, t**2, 'bs', label='y=x^2')
plt.plot(t, t**3, 'g^', label='y=x^3')
plt.legend()  # labelで指定した凡例を表示
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100
colors = np.random.rand(50)  # cf. Color map: [0, 1)の数値を色に対応づける仕組み
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.show()
```

```python
values = [1, 2, 3, 4, 5, 6]
x_pos = np.arange(len(values))

plt.bar(x_pos, values)
plt.show()
```

```python
values = [1, 2, 3, 4, 5, 6]
x_pos = np.arange(len(values))
objects = ['A', 'B', 'C', 'D', 'E', 'F']

plt.bar(x_pos, values)
plt.xticks(x_pos, objects)  # x軸のラベルをobjectsで置き換え
plt.show()
```

```python
rates = [40, 30, 10, 20]
labels = ['React', 'Vue', 'Angular', 'Svelte']
colors = ['blue', 'green', 'red', 'orange']

# 小数点以下1桁まで%表記
plt.pie(rates, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')  # 正円にする (plt.axis関数: 軸についての属性を設定する)
plt.show()
```
